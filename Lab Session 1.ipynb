```python
properti=pd.read_csv('/content/gdrive/MyDrive/SEMESTER 4/Data Mining/LabSession1/property.csv')
```

```python
#menampilkan data property
properti.info()
```

```python
#apakah pada data property tersebut terdapat kesalahan?
properti.isnull()
```

```python
#apakah pada data properti tersebut terdapat kesalahan namun dalam hitungan dijumlahkan total ?
properti.isnull().sum()
```

```python
#menampilkan atribut 'PID' pada properti dan ternyata terdapat eror pada index ke-4 sehingga 'NAN'
properti['PID']
```

```python
#menampilkan atribut 'PID' dengan perintah apakah data tersebut terdapat kesalahan? jawaban sesuai dengan index ke-4 sehingga 'true' karena terdapat kesalahan
properti['PID'].isnull()
```

```python
#mereplace pada data yang eror index ke-4 diubah menjadi 100005000.0 
properti['PID'].fillna(100005000.0, inplace=True)
```

```python
#menampilkan data pada atribut 'PID' yang sudah direplace
properti['PID']
```

```python
#menampilkan atribut 'ST_NUM' pada properti dan ternyata terdapat eror pada index ke-2 dan index ke-6 sehingga 'NAN'
properti['ST_NUM']
```

```python
#perintah untuk menampilkan nilai tengah,namun yang eror tidak dijumlahkan sehingga sehingga totalnya ada 7 dengan nilai tengahnya yaitu 203.0
median = properti['ST_NUM'].median()
```

```python
#perintah nilai index tersebut diubah dengan nilai tengahnya yaitu 203.0 
properti['ST_NUM'].fillna(median, inplace=True)
```

```python
#menampilkan atribut 'ST_NUM' yang sudah direplace
properti['ST_NUM']
```

```python
#nenampilkan data properti keseluruhan
properti
```

```python
#menampilkan atribut 'OWN_OCCUPIED' pada properti dan ternyata terdapat eror pada index ke-6 sehingga 'NAN'
properti['OWN_OCCUPIED']
```

```python
#nilai index eror tersebut direplace menjadi N 
properti['OWN_OCCUPIED'].fillna('N', inplace=True)
```

```python
#menampilkan atribut 'OWN_OCCUPIED' yang sudah direplace
properti['OWN_OCCUPIED']
```

```python
# pengisian berdasarkan angka
#perintah untuk mengubah angka(12) index ke-3 pada data diubah menjadi huruf 'Y'

properti.loc[3,'OWN_OCCUPIED'] = 'Y'
```

```python
#menampilkan data apda atribut'OWN_OCCUPIED' yang sudah direplace
properti['OWN_OCCUPIED']
```

```python
#menampilkan semua data properti yang sudah di replace
properti
```

```python
#menampilkan atribut 'NUM_BEDROOMS' pada properti dan ternyata terdapat eror pada index ke-2 dan index ke-5 sehingga 'NAN'
properti['NUM_BEDROOMS']
```

```python
#pad adalah data sebelum NAN sehingga pada atribut 'NUM_BEDROOMS'yaitu pada index ke-1 dan index ke-4
#pada nilai yang eror tersebut diubah menjadi  nilai diatas NAN yaitu '3'
properti['NUM_BEDROOMS'].fillna(method='pad', inplace=True)
```

```python
#menampilkan data apda atribut'NUM_BEDROOMS' yang sudah direplace
properti['NUM_BEDROOMS']
```

```python
#menampilkan atribut 'NUM_BATH' pada properti dan ternyata terdapat eror pada index ke-3 sehingga 'NAN'
properti['NUM_BATH']
```

```python
#bfill adalah data setelah NAN sehingga pada atribut 'NUM_MATH'yaitu pada index ke-3
#pada nilai yang eror tersebut diubah menjadi  nilai dibawah NAN yaitu '2'
properti['NUM_BATH'].fillna(method='bfill', inplace=True)
```

```python
#menampilkan data apda atribut'NUM_BATH' yang sudah direplace
properti['NUM_BATH']
```
